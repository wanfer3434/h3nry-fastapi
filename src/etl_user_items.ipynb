```python
import pandas as pd
import ast
import gzip

# Lista para almacenar los datos
rows = []

# Abre el archivo gzip y lo lee línea por línea
with gzip.open('/content/users_items.json.gz', 'rb') as f:
    for line in f:
        # Decodifica la línea y la evalúa con ast.literal_eval()
        data = ast.literal_eval(line.decode('utf-8'))
        rows.append(data)

# Crea el DataFrame a partir de la lista de datos
df_items = pd.DataFrame(rows)
```

```python
df_items
```

```python
#Unwrap 'items'
df_items1 = df_items.explode(['items'])
df_items2 = pd.json_normalize(df_items1['items']).set_index(df_items1['items'].index)
df_items = pd.concat([df_items2, df_items1], axis=1)
```

```python
df_items
```

```python
df_items.info()
```

```python
df_items = df_items.drop(['item_name', 'playtime_2weeks', 'items_count', 'steam_id', 'user_url','items', df_items.columns[-1]], axis=1)
```

```python
df_items
```

```python
# Remove duplicate rows based on the 'item_id', 'playtime_forever' column
df_items.dropna(subset=['item_id', 'playtime_forever'])

# Also remove empty or 0 rows
df_items[df_items['playtime_forever'] != 0]

# And any other potential data that we might have not seen
df_items.drop_duplicates()
df_items.dropna()
```
