```python
import pandas as pd
import ast
import gzip

# Lista para almacenar los datos del archivo JSON
row = []

# Ruta del archivo JSON comprimido
file_path = '/content/user_reviews.json.gz'

try:
    # Abrir el archivo JSON comprimido y leer línea por línea
    with gzip.open(file_path, 'rb') as f:
        for i in f:
            # Decodificar la línea y evaluarla con ast.literal_eval()
            line = i.decode("utf-8")
            row.append(ast.literal_eval(line))

    # Crear un DataFrame a partir de la lista de datos
    df_reviews = pd.DataFrame(row)

    # Mostrar el DataFrame
    #print(df_items.head())

except Exception as e:
    print("Error:", e)

```

En el Sistema de Análisis y Recomendación de Steam, el análisis de sentimientos desempeña un papel fundamental al comprender las opiniones y comentarios de los usuarios sobre los videojuegos. Este proceso, conocido como análisis de sentimientos, permite categorizar automáticamente las emociones expresadas en estas opiniones, proporcionando información valiosa para entender cómo se perciben los juegos por parte de la comunidad de usuarios.

El análisis de sentimientos se basa en la recopilación de comentarios y reseñas de los usuarios sobre los juegos que han jugado. Estos comentarios se someten a un proceso automatizado en el que un algoritmo especializado evalúa el tono emocional de cada comentario. Este algoritmo asigna un "puntaje de sentimiento" a cada comentario, indicando si el usuario expresó sentimientos positivos, negativos o neutros hacia el juego.

Este proceso es esencial para comprender la satisfacción del usuario, identificar tendencias en las opiniones de los juegos y mejorar la calidad de las recomendaciones. Al analizar el sentimiento general de la comunidad hacia un juego específico, podemos ofrecer recomendaciones más precisas y personalizadas a los usuarios, ayudándoles a descubrir nuevos juegos que se adapten a sus preferencias y gustos individuales.

```python
import nltk
# descarga del diccionario vader_lexicon
nltk.download('vader_lexicon')
# importar el analizador de sentimientos vader
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Importar la clase SentimentIntensityAnalyzer desde la biblioteca nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Instanciar el analizador de sentimientos
sia = SentimentIntensityAnalyzer()

# Define una función para categorizar los sentimientos
def categorize_sentiment(score):
    if score < -0.05:
        return 0  # Negativo
    elif score > 0.05:
        return 2  # Positivo
    else:
        return 1  # Neutral

# Asegurarse de que la columna 'review' sea de tipo str
df_reviews['review'] = df_reviews['review'].astype(str)

# Aplicar el análisis de sentimientos a las reseñas y categorizar los resultados
df_reviews['review'] = df_reviews['review'].apply(lambda review: sia.polarity_scores(review)['compound'])
df_reviews['review'] = df_reviews['review'].map(categorize_sentiment)
```
