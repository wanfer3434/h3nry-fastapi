```python
import pandas as pd
import json

file_path = r'/content/output_steam_games.json'  # Replace with your file path

data = []
with open(file_path, 'r') as file:
    for line in file:
        try:
            json_object = json.loads(line)
            data.append(json_object)
        except json.JSONDecodeError:
            # You can handle or log errors here if needed
            pass

df = pd.DataFrame(data)

# Optionally, check the first few rows of the DataFrame
print(df.head())
df
```

```python
df.dropna(how='all', inplace=True)
```

```python
df
```

```python
columns_to_drop = ['publisher', 'app_name', 'url', 'reviews_url', 'specs', 'price', 'early_access']
# Eliminar columnas y restablecer el índice en una sola línea
df = df.drop(columns=columns_to_drop).reset_index(drop=True)
```

```python
df
```

```python
df_exploded = df.explode('genres')
```

```python
df_exploded
```

```python
cantidad_genero_unico = df_exploded['genres'].nunique()
frecuencia_genero_unico = df_exploded['genres'].value_counts()

print(f"Total cantidad_genero_unico: {cantidad_genero_unico}")
print("frecuencia_genero_unico:\n", frecuencia_genero_unico)
```

```python
df.to_csv('steam_games.csv', index=False)
```

```python
import ast
import pandas as pd

# Function to safely parse a string list
def parse_string_list(string_list):
    try:
        return ast.literal_eval(string_list)
    except (ValueError, SyntaxError):
        return []

# Update missing genres based on tags using apply and lambda function
df['genres'] = df.apply(lambda row: row['genres'] if pd.notna(row['genres']) else parse_string_list(row['tags']), axis=1)
```

```python
missing_genres = df['genres'].isna().any()

if missing_genres:
    print("Hay datos faltantes en la columna 'genres'.")
else:
    print("No hay datos faltantes en la columna 'genres'.")
```

```python
df = df.drop(columns=['tags'])
```

```python
df
```

```python
df = df.dropna(subset=['release_date'], inplace=True)
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
df = df.dropna(subset=['release_date'])
df['release_date'] = df['release_date'].astype(int)
df.rename(columns={'release_date': 'release_year'}, inplace=True)

```

```python
df
```

```python
# Comprobamos si hay duplicados en 'id' y datos faltantes

df = (df.drop_duplicates(subset='id')
        .dropna(subset=['id'])
        .astype({'id': int})
        .rename(columns={'id': 'game_id'}))
```
